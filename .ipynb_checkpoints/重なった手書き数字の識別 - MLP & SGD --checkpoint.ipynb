# 重なった手書き数字の識別 - MLP & SGD -

## 0.ライブラリのimport

```python
import csv
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.model_selection import train_test_split

np.random.seed(seed=0)
```

## 1.datasetの準備

```python
# MNISTの画像とラベルの準備
if os.path.exists('mnist_784'):
    with open('mnist_784','rb') as f:
        mnist = pickle.load(f)
else:
    mnist = datasets.fetch_openml('mnist_784')
    with open('mnist_784', 'wb') as f:
        pickle.dump(mnist, f)
    
X, T = mnist.data, mnist.target.astype(np.int32)
```

```python
# 2つの画像を重ね合わせる。100次元(→55次元にもできる)クラスの分類。
n = len(X) // 2
T_100 = np.zeros((n, 100))
for i in range(n):
    X[i] = np.maximum(X[i], X[i + n])
    max_i, min_i = max(T[i], T[i + n]), min(T[i], T[i + n])
    idx = min_i * 10 + max_i
    T_100[i][idx] = 1
    
X = X[:n]
T = T_100
```

```python
# train test に分割
X_train, X_test, T_train, T_test = train_test_split(X, T, test_size=0.2)
```

```python
# サイズの確認。
print(X_train.shape)
print(X_test.shape)
print(T_train.shape)
print(T_test.shape)
```

```python
# サンプル画像を表示する。2つの画像が重なっている。
plt.gray()
for i in range(5):
    plt.imshow(X_train[i].reshape(28, 28))
    plt.show()
    print('Label: ', np.where(T_train[i] == 1))
```

## 2.関数の定義

```python
def softmax(x):
    x = x.T
    _x = x - np.max(x, axis=0)
    _x = np.exp(_x) / np.sum(np.exp(_x), axis=0)
    return _x.T
```

```python
def cross_entropy_error(t, y):
    delta = 1e-8
    error = -np.mean(np.sum(t * np.log(y + delta), axis=1))
    return error
```

```python
class SoftmaxCrossEntropyLoss():
    def __init__(self):
        self.y = None
        self.t = None
        self.loss = None
        
    def __call__(self, t, y):
        self.y = softmax(y)
        self.t = t.copy()
        self.loss = cross_entropy_error(self.t, self.y)
        return self.loss
    
    def backward(self):
        batch_size = self.t.shape[0]
        dy = self.y - self.t
        dy /= batch_size
        return dy
```

```python
class ReLU():
    def __init__(self):
        self.mask = None

    def __call__(self, x):
        self.mask = (x <= 0)
        out = x.copy()
        out[self.mask] = 0 
        return out

    def backward(self, dout):
        dout[self.mask] = 0
        dx = dout

        return dx
```

```python
class FullyConnectedLayer():
    def __init__(self, input_shape, output_shape):
        self.w = np.random.randn(input_shape, output_shape) * 0.01
        self.b = np.zeros(output_shape, dtype=np.float)
        self.x = None
        self.dw = None
        self.db = None
        
    def __call__(self, x):
        self.x = x
        out = np.dot(x, self.w) + self.b
        return out
    
    def backward(self, dout):
        dx = np.dot(dout, self.w.transpose())
        batch_size = dx.shape[0]
        self.dw = np.dot(self.x.transpose(), dout)
        self.db = np.sum(dout, axis=0)
        return dx
```

## 3.モデルの作成

```python
class MLP_classifier():

    def __init__(self):

        self.fc1 = FullyConnectedLayer(784, 256)
        self.relu1 = ReLU()
        self.fc2 = FullyConnectedLayer(256, 256)
        self.relu2 = ReLU()
        self.fc3 = FullyConnectedLayer(256, 100)
        self.out = None

        self.criterion = SoftmaxCrossEntropyLoss()

    def forward(self, x):
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        self.out = self.fc3(x)
        
        return self.out

    def backward(self, t):
        loss = self.criterion(t, self.out)
        d = self.criterion.backward()
        d = self.fc3.backward(d)
        d = self.relu2.backward(d)
        d = self.fc2.backward(d)
        d = self.relu1.backward(d)
        d = self.fc1.backward(d)
        
        return loss

    def optimize_GradientDecent(self, lr):
        for fc in [self.fc1, self.fc2, self.fc3]:
            fc.w -= lr * fc.dw
            fc.b -= lr * fc.db
```

## 4.学習

```python
train_loss_list = []
train_acc_list = []
test_loss_list = []
test_acc_list = []
```

```python
model = MLP_classifier()

lr = 0.01
n_epoch = 100

for n in range(n_epoch):
    y = model.forward(X_train)
    train_loss = model.backward(T_train)
    train_acc = np.mean(np.equal(np.argmax(y, axis=1), np.argmax(T_train, axis=1)))
    model.optimize_GradientDecent(lr)
    
    y = model.forward(X_test)
    test_loss = model.backward(T_test)
    test_acc = np.mean(np.equal(np.argmax(y, axis=1), np.argmax(T_test, axis=1)))
    
    train_loss_list.append(train_loss)
    train_acc_list.append(train_acc)
    test_loss_list.append(test_loss)
    test_acc_list.append(test_acc)
    
    if n % 10 == 0:
        print(f'EPOCH {n + 1} | train_loss {train_loss:.5f} | test_loss {test_loss:.5f} | test_acc {test_acc:.2%}')
```

## 5.学習結果の表示

```python
x = range(n_epoch)
plt.plot(x, train_loss_list, color='blue', linestyle='-')
plt.plot(x, test_loss_list, color='green', linestyle='--')

plt.legend(['train_loss', 'test_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.grid(True)
plt.show()
```

```python
x = range(n_epoch)
plt.plot(x, train_acc_list, color='blue', linestyle='-')
plt.plot(x, test_acc_list, color='green', linestyle='--')

plt.legend(['train_acc', 'test_acc'])
plt.xlabel('epoch')
plt.ylabel('acc')
plt.grid(True)
plt.show()
```
